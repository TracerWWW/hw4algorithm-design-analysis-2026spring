2. 算法功能及是否最高效

该算法的功能是对数组 $A$ 中从下标 $p$ 到 $r$ 的元素进行排序。

具体来说，它可以将数组按非降序排列，即：

$$
A[p]\le A[p+1]\le \cdots \le A[r]
$$

该算法的思想是：

1. 如果区间中只有两个元素，就直接比较并交换；
2. 如果区间中有三个或更多元素，就先排序前 $\frac{2}{3}$ 部分；
3. 再排序后 $\frac{2}{3}$ 部分；
4. 最后再次排序前 $\frac{2}{3}$ 部分。

这个过程能够保证整个区间最终有序。

但是，该算法不是最高效的排序算法。

它的时间复杂度为：

$$
\Theta\left(n^{\log_{3/2}3}\right)
\approx
\Theta(n^{2.71})
$$

而常见的高效比较排序算法，例如归并排序和堆排序，时间复杂度为：

$$
\Theta(n\log n)
$$

快速排序的平均时间复杂度也是：

$$
\Theta(n\log n)
$$

因此，该算法比常见高效排序算法慢得多。



````python
#使用归并排序，时间复杂度O(n log n)
import math

def merge_sort(A):
    if len(A) <= 1:
        return A

    mid = len(A) // 2
    left = merge_sort(A[:mid])
    right = merge_sort(A[mid:])

    return merge(left, right)


def merge(left, right):
    result = []
    i = j = 0

    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    result.extend(left[i:])
    result.extend(right[j:])
    return result


A = [5, 3, 1, 4, 2]
print(merge_sort(A))
````

## 问题 3

给定一个包含n个元素的数组 `profits` , 它的第 `i` 个元素 `profits[i]` 表示一支股票第 `i` 天的**收益**（正数表示涨, 负数表示跌）。你只能选择 **某一天** 买入这只股票, 并选择在 **未来的某一个不同的日子** 卖出该股票。

1. 设计一个算法来计算你所能获取的最大利润和对应买入和卖出的日期。请分析算法方案, 计算其时间复杂度, 并且使用python编程实现该算法。

2. \* 设计一个时间复杂度为 $O(n)$的算法实现该算法

e.g. :
---
profits=[3,2,1,-7,5,2,-1,3,-1], 第5天买入, 第8天卖出, 收益最大：9

idea:
算法从左到右扫描数组，记录当前为止出现过的最低价格以及对应日期。
对于每一天，假设当天卖出，则最大收益为当天价格减去此前最低价格。
如果该收益大于当前最大收益，则更新答案。
最后得到最大收益及对应买入、卖出日期。

该算法只需要遍历数组一次，因此时间复杂度为 O(n)，只使用常数个变量，因此空间复杂度为 O(1)。


````python
def max_profit_linear(profits):
    if len(profits) < 2:
        return 0, -1, -1

    min_price = profits[0]
    min_day = 0

    max_profit = 0
    buy_day = -1
    sell_day = -1

    for i in range(1, len(profits)):
        current_profit = profits[i] - min_price

        if current_profit > max_profit:
            max_profit = current_profit
            buy_day = min_day
            sell_day = i

        if profits[i] < min_price:
            min_price = profits[i]
            min_day = i

    return max_profit, buy_day, sell_day


profits = [3, 2, 1, -7, 5, 2, -1, 3, -1]
ans = max_profit_linear(profits)

print("最大收益:", ans[0])
print("买入日期:", ans[1] + 1)
print("卖出日期:", ans[2] + 1)

# your algorithm time complexity is:O(n)
# your algorithm space complexity is:O(1)
````

## 问题 4

给定一个包含不同整数元素的数组 $ A[1..n]$ ,并且满足条件：$A[1]>A[2]$ 并且 $A[n-1]<A[n]$; 规定：如果一个元素比它两边的邻居元素都小, 即：$A[x]<A[x-1], A[x]<A[x+1]$ , 称这个元素A[x]为“局部最小”。通过遍历一次数组, 我们可以很容易在 $O(n)$的时间复杂度下找到一个局部最小值, 


1. 分析该问题, 设计一个算法在$O(logn)$的时间复杂度下找到一个局部最小(返回数值), 要求：分析算法设计思路, 并且使用python编程实现
2. \* 设计算法找出所有局部最小值, 分析算法设计思路, 并使用python编程实现

e.g.:
---
A=[9, 3, 7, 2, 1, 4, 5 ] 时,  局部最小元素为 3, 1

idea:
可以使用二分查找。
每次取中间位置 mid：
如果 A[mid] 同时小于左右相邻元素，那么 A[mid] 就是局部最小值。
如果 A[mid] > A[mid-1]，说明左边方向是下降的，因此左半部分一定存在局部最小值。
如果 A[mid] > A[mid+1]，说明右边方向是下降的，因此右半部分一定存在局部最小值。
由于每次都能排除一半区间，所以时间复杂度为 O(log n)。

````python
def find_one_local_min(A):
    n = len(A)

    left = 1
    right = n - 2

    while left <= right:
        mid = (left + right) // 2

        if A[mid] < A[mid - 1] and A[mid] < A[mid + 1]:
            return A[mid]

        elif A[mid] > A[mid - 1]:
            right = mid - 1

        else:
            left = mid + 1

    return None


A = [9, 3, 7, 2, 1, 4, 5]
print(find_one_local_min(A))

# your algorithm time complexity is:O(log n)
````

````python
def find_all_local_min(A):
    result = []

    for i in range(1, len(A) - 1):
        if A[i] < A[i - 1] and A[i] < A[i + 1]:
            result.append(A[i])

    return result


A = [9, 3, 7, 2, 1, 4, 5]
print(find_all_local_min(A))

# 算法设计思路：

# 要找出所有局部最小值，必须检查数组中每一个可能成为局部最小值的位置。
# 根据题目定义，元素 A[x] 是局部最小值，当且仅当它同时满足：
# A[x] < A[x-1]
# 并且：
# A[x] < A[x+1]
# 因此，可以从数组的第二个元素开始，遍历到倒数第二个元素。对于每个位置 i，判断它是否同时小于左右两个相邻元素。如果满足条件，就把 A[i] 加入结果数组。
# 由于题目已经给出条件：
# A[1] > A[2]
# A[n-1] < A[n]
# 所以数组的局部最小值一定出现在中间某个位置。遍历所有中间位置，就可以找出所有局部最小值。
````

````python
#1. 排序方法
import numpy as np
def k_smallest_by_sort(arr, k):
    arr_sorted = sorted(arr)
    return arr_sorted[:k]


arr = [5, 4, 3, 2, 6, 1, 88, 33, 22, 107]
k = 3

print(k_smallest_by_sort(arr, k))

#时间复杂度： O(n log n)
#因为排序需要 O(n log n)，取前 k 个需要 O(k)，总复杂度为： O(n log n)
````

````python
import heapq

def k_smallest_by_min_heap(arr, k):
    heap = arr[:]
    heapq.heapify(heap)

    result = []
    for _ in range(k):
        result.append(heapq.heappop(heap))

    return result


arr = [5, 4, 3, 2, 6, 1, 88, 33, 22, 107]
k = 3

print(k_smallest_by_min_heap(arr, k))

# your algorithm time complexity is:O(n + k log n)
# heapify 建堆需要 O(n)
# 每次 Extract-min 需要 O(log n)
# 执行 k 次需要 O(k log n)
````

````python
import heapq

def k_smallest_stream(nums, k):
    max_heap = []

    for x in nums:
        if len(max_heap) < k:
            heapq.heappush(max_heap, -x)
        else:
            if x < -max_heap[0]:
                heapq.heapreplace(max_heap, -x)

    result = [-x for x in max_heap]
    result.sort()
    return result


arr = [5, 4, 3, 2, 6, 1, 88, 33, 22, 107]
k = 3

print(k_smallest_stream(arr, k))

# 先把前 k 个数放入最大堆。
# 之后每读入一个新数 x，就和堆顶最大值比较。
# 如果 x 比堆顶小，说明 x 应该进入最小的 k 个数，于是删除堆顶，再加入 x。
# 如果 x 比堆顶大，则忽略 x。

# 时间复杂度： O(n log k)
# 空间复杂度： O(k)
````

## 问题 6

> 星球碰撞问题：现有n个星球，在同一条直线上运行，如数组A所示，元素的绝对值表示星球的质量，负数表示星球自右向左运动，正数表示星球自左向右运动，当两个星球相撞的时候，质量小的会消失，大的保持不变，**质量相同的两个星球碰撞后自右向左运动的星球消失，自左向右的星球保持不变**，假设所有星球的速度大小相同。
>
> $ A=[23,-8, 9, -3, -7, 9, -23, 22] $

请设计一个算法模拟星球的运行情况，输出最终的星球存续情况（输出一个数组），分析算法设计思路，计算时间复杂度，并基于python编程实现。

e.g.
---
1.  输入： A=[-3,-6,2,8, 5,-8,9,-2,1]， 输出：[-3, -6, 2, 8, 9, 1]

2. 输入：A=[23,-8, 9, -3, -7, 9, -23, 22], 输出：[23, 22]


算法设计思路

使用一个栈 $S$ 来保存当前仍然存活的星球。

从左到右遍历数组 $A$ 中的每个星球 $x$。

如果 $x>0$，说明它向右运动，暂时不会和栈中已有星球发生碰撞，直接入栈。

如果 $x<0$，说明它向左运动，只有当栈顶星球向右运动时，才可能发生碰撞。也就是当：

$$
S[-1]>0
$$

且：

$$
x<0
$$

时，需要不断比较二者质量。

设当前向左运动星球质量为：

$$
|x|
$$

栈顶向右运动星球质量为：

$$
S[-1]
$$

分三种情况：

1. 若 $S[-1]<|x|$，则栈顶星球质量更小，栈顶星球消失，继续与更前面的星球比较；
2. 若 $S[-1]=|x|$，则两个星球质量相同，且自右向左运动的星球消失，因此当前星球 $x$ 消失，栈顶星球保留；
3. 若 $S[-1]>|x|$，则当前星球 $x$ 消失，栈顶星球保留。

遍历结束后，栈中剩余的元素就是最终存活的星球。

---

时间复杂度分析

每个星球最多入栈一次、出栈一次。

因此总时间复杂度为：

$$
O(n)
$$

其中 $n$ 是数组 $A$ 的长度。

栈中最多保存 $n$ 个星球，因此空间复杂度为：

$$
O(n)
$$

---


````python

def asteroid_collision(A):
    stack = []

    for x in A:
        alive = True

        while alive and x < 0 and stack and stack[-1] > 0:
            if stack[-1] < -x:
                stack.pop()
            elif stack[-1] == -x:
                alive = False
                break
            else:
                alive = False
                break

        if alive:
            stack.append(x)

    return stack


A1 = [-3, -6, 2, 8, 5, -8, 9, -2, 1]
A2 = [23, -8, 9, -3, -7, 9, -23, 22]

print(asteroid_collision(A1))
print(asteroid_collision(A2))
# 时间复杂度： O(n)
# 空间复杂度： O(n) 最坏情况下，所有小行星都不会碰撞，堆栈中会存储所有小行星。
````

## 问题 7

> 0-1背包问题：现在有4块大理石，每个大理石对应的重量和价值使用一个元组表示, 即（重量，价值），4块大理石的重量和价值为：[(5,10), (4,40), (6,30), (3, 50)], 假设你有一辆最大承重为9的小推车，请问使用这个小推车装大理石的最大价值为多少

请设计一个算法解决该问题，分析算法设计思路，计算时间复杂度，并基于python编程实现

idea:
设 dp[i][w] 表示只考虑前 i 个物品，背包容量为 w 时可以获得的最大价值。
对于第 i 个物品，有两种选择：
1. 不选它：
   dp[i][w] = dp[i-1][w]
2. 选它：
   dp[i][w] = dp[i-1][w-weight] + value
所以状态转移方程为：
dp[i][w] = max(dp[i-1][w], dp[i-1][w-weight] + value)

````python
def knapsack_01(items, capacity):
    n = len(items)

    dp = [[0] * (capacity + 1) for _ in range(n + 1)]

    for i in range(1, n + 1):
        weight, value = items[i - 1]

        for w in range(capacity + 1):
            if weight > w:
                dp[i][w] = dp[i - 1][w]
            else:
                dp[i][w] = max(
                    dp[i - 1][w],
                    dp[i - 1][w - weight] + value
                )

    return dp[n][capacity]


items = [(5, 10), (4, 40), (6, 30), (3, 50)]
capacity = 9

print(knapsack_01(items, capacity))

# your algorithm time complexity is:O(nW)
````

给你二叉搜索树的根节点 root ，该树中的恰好两个节点的值被错误地交换。请在不改变其结构的情况下，恢复这棵树 。设计算法解决该问题，分析算法设计思路，计算时间复杂度，并基于python编程实现

e.g.:
---
1.

<div style="text-align:center;width:200px;"> <img alt="" src="./fig/hw3q9-1.jpg"> </div>
输入：root = [1,3,null,null,2]， 输出为：[3,1,null,null,2]， 

解释：3 不能是 1 的左孩子，因为 3 > 1 。交换 1 和 3 使二叉搜索树有效。

2.

 <div style="text-align:center;width:200px;"> <img alt="" src="./fig/hw3q9-1.jpg"> </div>
输入： root=[3,1,4,null,null,2]， 输出为：[2,1,4,null,null,3]

解释：2 不能在 3 的右子树中，因为 2 < 3 。交换 2 和 3 使二叉搜索树有效。

idea:
对二叉搜索树进行中序遍历。
正常情况下，中序遍历得到的节点值应该从小到大排列。
如果发现 prev.val > 当前节点值，说明出现了错误顺序。
第一次出现错误时，把 prev 记为 first，把当前节点记为 second。
如果后面再次出现错误，更新 second 为当前节点。
最后交换 first 和 second 的值。

举例：
root = [1,3,null,null,2]

中序遍历结果是：
3, 2, 1

发现两处逆序：
3 > 2 和
2 > 1
所以交换 3 和 1，结果为：
[3,1,null,null,2]

时间复杂度：
O(n)
因为每个节点只遍历一次。

空间复杂度：
O(h)
其中 h 是树的高度，递归调用栈最多占用树高的空间。最坏情况下树退化成链表，空间复杂度为 O(n)；如果是平衡树，则为 O(log n)。

````python
from collections import deque


class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


def build_tree(values):
    if not values or values[0] is None:
        return None

    root = TreeNode(values[0])
    queue = deque([root])
    i = 1

    while queue and i < len(values):
        node = queue.popleft()

        if i < len(values) and values[i] is not None:
            node.left = TreeNode(values[i])
            queue.append(node.left)
        i += 1

        if i < len(values) and values[i] is not None:
            node.right = TreeNode(values[i])
            queue.append(node.right)
        i += 1

    return root


def tree_to_list(root):
    if root is None:
        return []

    result = []
    queue = deque([root])

    while queue:
        node = queue.popleft()

        if node is None:
            result.append(None)
        else:
            result.append(node.val)
            queue.append(node.left)
            queue.append(node.right)

    while result and result[-1] is None:
        result.pop()

    return result


def recover_tree(root):
    first = None
    second = None
    prev = None

    def inorder(node):
        nonlocal first, second, prev

        if node is None:
            return

        inorder(node.left)

        if prev is not None and prev.val > node.val:
            if first is None:
                first = prev
            second = node

        prev = node

        inorder(node.right)

    inorder(root)

    if first is not None and second is not None:
        first.val, second.val = second.val, first.val


root = build_tree([1, 3, None, None, 2])
recover_tree(root)
print(tree_to_list(root))


# your algorithm time complexity is:O(n)
````

## 问题 9 \*
给定 `n` 个非负整数表示每个宽度为 `1` 的柱子的高度图，计算按此排列的柱子，下雨之后能接多少雨水。



<div style="text-align:center;width:800px;"> <img alt="" src="./fig/hw3q6.png"> </div>

请分析算法设计思路，计算时间复杂度，并基于python编程实现

idea:

核心思想：某个位置 i 能接多少水，取决于它左边最高柱子和右边最高柱子中较矮的那个。
公式： water[i] = min(left_max[i], right_max[i]) - height[i]
使用两个指针 left 和 right，分别从数组两端向中间移动。
同时维护 left_max 和 right_max，表示当前左边和右边遇到过的最高柱子。
如果 height[left] < height[right]，说明左边当前位置能接多少水只取决于 left_max。
如果 height[left] >= left_max，则更新 left_max。
否则当前位置能接的水为 left_max - height[left]。
如果 height[right] <= height[left]，同理处理右边。
最终累加得到总雨水量。

````python
def trap(height):
    if not height:
        return 0

    left = 0
    right = len(height) - 1

    left_max = 0
    right_max = 0

    water = 0

    while left < right:
        if height[left] < height[right]:
            if height[left] >= left_max:
                left_max = height[left]
            else:
                water += left_max - height[left]
            left += 1
        else:
            if height[right] >= right_max:
                right_max = height[right]
            else:
                water += right_max - height[right]
            right -= 1

    return water


height1 = [0, 1, 0, 2, 1, 0, 1, 3, 2, 1, 2, 1]
height2 = [4, 2, 0, 3, 2, 5]

print(trap(height1))
print(trap(height2))

# your algorithm time complexity is:O(n)

````

## 问题 10

给你一个二叉树的根节点 root ，树中每个节点都存放有一个 0 到 9 之间的数字。每条从根节点到叶节点的路径都代表一个数字：

> 例如，从根节点到叶节点的路径 1 -> 2 -> 3 表示数字 123 。

设计一个算法计算从根节点到叶节点生成的所有数字之和 ，分析算法设计思路，计算时间复杂度，并基于python编程实现

叶节点:是指没有子节点的节点。

e.g.:
---
1.

 <div style="text-align:center;width:200px;"> <img alt="figure" src="./fig/hw3q10-1.jpg"> </div>

 输入：root = [1,2,3], 输出：25

解释：从根到叶子节点路径 1->2 代表数字 12，1->3 代表数字 13，因此和为 12 + 13 = 25 。

2.

 <div style="text-align:center;width:200px;"> <img alt="figure" src="./fig/hw3q10-2.jpg"> </div>

 输入：root = [4,9,0,5,1], 输出：1026

解释：从根到叶子节点路径 4->9->5 代表数字 495，4->9->1 代表数字 491，4->0 代表数字 40。因此和为 495 + 491 + 40 = 1026 。

idea:
从根节点开始进行深度优先搜索。
用 current 记录从根节点到当前节点形成的数字。
每访问一个节点，就更新 current = current * 10 + node.val。
如果当前节点是叶子节点，说明形成了一个完整数字，将其返回。
如果不是叶子节点，则继续递归左右子树，并把左右子树结果相加。

````python
from collections import deque


class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


def build_tree(values):
    if not values or values[0] is None:
        return None

    root = TreeNode(values[0])
    queue = deque([root])
    i = 1

    while queue and i < len(values):
        node = queue.popleft()

        if i < len(values) and values[i] is not None:
            node.left = TreeNode(values[i])
            queue.append(node.left)
        i += 1

        if i < len(values) and values[i] is not None:
            node.right = TreeNode(values[i])
            queue.append(node.right)
        i += 1

    return root


def sum_numbers(root):
    def dfs(node, current):
        if node is None:
            return 0

        current = current * 10 + node.val

        if node.left is None and node.right is None:
            return current

        return dfs(node.left, current) + dfs(node.right, current)

    return dfs(root, 0)


root1 = build_tree([1, 2, 3])
print(sum_numbers(root1))

root2 = build_tree([4, 9, 0, 5, 1])
print(sum_numbers(root2))

# your algorithm time complexity is:O(n)，因为每个节点只访问一次。
````
